## Feature Informations
* People
    * ID: Customer unique identifier
    * Year_birth : Customer's birth year

    * Marital_status : Customer's marital status
    * income : Customer's yearly household income 
    * kidhome : Number of children in customer household 
    * Teenhome : Number of teenagers in customer's household
    * Dt_customer : Date of customer's enrollement with the company
    * recency : Number of days since customer's last purchase
    * Complain : 1 if the customer complained in the last 2 years,0 otherwise

* Products 
    - MnWines: Amount spent on the wine last 2 years 
    - MntFruits : Amount spent on fruits in last 2 years
    - MntMeatproducts : Amount spent on meat in last 2 years
    - MntFishProducts : Amount spent on the Fish in last 2 years
    - MntSweetProducts : Amount spent on sweet in last 2 years
    - MntGoldprods : Amount spent on the gold in last 2 years

* Promotion

  -  NumDealsPurchases: Number of purchases made with a discount
  -  AcceptedCmp1: 1 if customer accepted the offer in the 1st campaign, 0 otherwise
  -  AcceptedCmp2: 1 if customer accepted the offer in the 2nd campaign, 0 otherwise
  -  AcceptedCmp3: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise
  -  AcceptedCmp4: 1 if customer accepted the offer in the 4th campaign, 0 otherwise
  -  AcceptedCmp5: 1 if customer accepted the offer in the 5th campaign, 0 otherwise
  -  Response: 1 if customer accepted the offer in the last campaign, 0 otherwise
* Place
  -   NumWebPurchases: Number of purchases made through the companyâ€™s website
  -   NumCatalogPurchases: Number of purchases made using a catalogue
  -   NumStorePurchases: Number of purchases made directly in stores
  -   NumWebVisitsMonth: Number of visits to companyâ€™s website in the last month

# Feature Extractions-
- Creating feature that would describe the profile of the customer

```python
from datetime import datetime


df_raw=df.copy()

df_raw["Age"] = 2024 - df_raw['Year_Birth'] ## creating column to store age of the customer

#replacing the education level to numeric values
df_raw["Education"].replace({"Basic":0,"2n Cycle":1,"Graduation":2,"Master":3,"PhD":4}, inplace=True)

# Replacing the marital status with numberic values
df_raw["Marital_Status"].replace({"Married":1,"Together":2,"Absurd":0,"Widow":0,"YOLO":0, "Divorced":0, "Single":0,"Alone":0},inplace=True)

# creating column to store number of children in the household
df_raw["Children"]=df_raw["Kidhome"]+df_raw["Teenhome"]

# creating the column to store family size
df_raw["Family_size"]=df_raw["Marital_Status"]+df_raw['Children']+1


#creating new field to store the total amount spent by the customer and promo used
df_raw['Total_spending']=df_raw['MntWines']+df_raw['MntFruits']+df_raw['MntMeatProducts']+df_raw['MntFishProducts']+df_raw['MntSweetProducts']+df_raw['MntGoldProds']
df_raw['Total_promo']=df['AcceptedCmp1']+df['AcceptedCmp2']+df['AcceptedCmp3']+df['AcceptedCmp4']+df['AcceptedCmp5']


#to calculate how long the customer is with the Store
df_raw['Dt_Customer'] = pd.to_datetime(df_raw["Dt_Customer"])
today = datetime.today()
df_raw['Days_as_Customer'] = (today - df_raw['Dt_Customer']).dt.days
df_raw['Offers_Responded_To'] = (df_raw['AcceptedCmp1'] + df_raw['AcceptedCmp2'] + df_raw['AcceptedCmp3'] +
                                 df_raw['AcceptedCmp4'] + df_raw['AcceptedCmp5'] + df_raw['Response'])
df_raw["Parental Status"] = np.where(df_raw["Children"] > 0, 1, 0)


#dropping columns which are already used to create new features
columns_to_drop = ['Year_Birth','Kidhome','Teenhome']
df_raw.drop(columns = columns_to_drop, axis = 1, inplace=True)
df_raw.rename(columns={"Marital_Status": "Marital Status","MntWines": "Wines","MntFruits":"Fruits",
                   "MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets",
                   "MntGoldProds":"Gold","NumWebPurchases": "Web","NumCatalogPurchases":"Catalog",
                   "NumStorePurchases":"Store","NumDealsPurchases":"Discount Purchases"},
          inplace = True)

df = df_raw[["Age","Education","Marital Status","Parental Status","Children","Income","Total_spending","Days_as_Customer","Recency","Wines","Fruits","Meat","Fish","Sweets","Gold","Web","Catalog","Store","Discount Purchases","Total_promo","NumWebVisitsMonth"]]
```
